# Regresión Lineal Simple



La regresión lineal es uno de los modelos más fundamentales y ampliamente utilizados en el ámbito del Machine Learning y la estadística. Se trata de una técnica que nos permite entender y modelar la relación entre una variable dependiente (o variable de respuesta) y una o más variables independientes (o variables explicativas). Esta relación se representa mediante una línea recta que busca aproximarse lo mejor posible a los datos observados, minimizando el error entre las predicciones y los valores reales.

El modelo de regresión lineal se basa en la suposición de que existe una relación lineal entre las variables, es decir, que el cambio en la variable dependiente es proporcional al cambio en la variable independiente. En su forma más básica, la regresión lineal simple se expresa como:

$$
y=b_0+b_1*x_1
$$

donde:

- **y** es la variable dependiente que deseamos predecir.
- **x** es la variable independiente que utilizamos para hacer la predicción.
- **$b_0$ es la intersección con el eje y, o el valor de y cuando x es igual a cero.**
- **$b_1$** es la pendiente de la línea, que indica la tasa de cambio de y con respecto a x.
- **ε** es el término de error, que representa la diferencia entre el valor predicho y el valor real debido a factores no considerados en el modelo.

En el contexto del Machine Learning, la regresión lineal se utiliza para realizar predicciones numéricas y también como una herramienta inicial para explorar relaciones entre variables. Es un modelo interpretable y fácil de implementar, lo cual lo convierte en una elección ideal para quienes se inician en el campo del aprendizaje automático. A través del uso de algoritmos como el método de mínimos cuadrados, se ajusta la línea que minimiza la suma de los cuadrados de los errores, proporcionando una solución óptima para el conjunto de datos disponible.

Sin embargo, la regresión lineal tiene ciertas limitaciones. La principal es la suposición de linealidad, que no siempre se cumple en situaciones más complejas. Además, es sensible a los valores atípicos, ya que unos pocos datos extremos pueden distorsionar la línea ajustada. Por estas razones, en problemas más sofisticados, se suelen utilizar modelos más avanzados, como las redes neuronales en el caso del Deep Learning, que pueden capturar patrones no lineales.

A pesar de sus limitaciones, la regresión lineal sigue siendo una herramienta fundamental para entender conceptos básicos de predicción y para desarrollar intuición sobre el comportamiento de los datos. Su simplicidad e interpretabilidad hacen que sea el punto de partida perfecto para adentrarse en el mundo del Machine Learning.
## Tipos de Variables:
En regresión lineal simple, las variables pueden clasificarse principalmente en dos tipos: variables numéricas y variables categóricas. Cada una tiene un papel específico y, dependiendo del tipo de variable independiente, el enfoque para construir el modelo de regresión puede variar considerablemente.

### Variables Numéricas
Las variables numéricas, también conocidas como variables cuantitativas, son aquellas que representan cantidades medibles y que tienen un valor continuo o discreto. Ejemplos comunes de variables numéricas incluyen el ingreso mensual, la temperatura, la edad, el precio de un producto, entre otros. En el contexto de la regresión lineal, estas variables se utilizan para determinar la relación entre una variable dependiente y la variable independiente de manera directa. La regresión lineal simple trata de ajustar una línea a estos puntos numéricos para hacer predicciones sobre nuevas observaciones.

### Variables Categóricas
Las variables categóricas, o cualitativas, son aquellas que no representan cantidades numéricas sino que indican grupos o categorías. Por ejemplo, el color de un coche (rojo, azul, verde), el tipo de ciudad (urbana, rural), o el género de una persona (masculino, femenino) son variables categóricas. Estas variables deben ser transformadas antes de ser utilizadas en un modelo de regresión lineal, ya que los algoritmos de regresión requieren valores numéricos. Un método común para esta transformación es la codificación de variables categóricas mediante "dummy variables" o variables ficticias, donde cada categoría se representa con un valor binario (0 o 1).

### Ejemplo del Uso de Variables en Regresión Lineal
Imaginemos que queremos predecir el precio de un coche. La variable dependiente sería el precio, mientras que algunas de las variables independientes podrían ser tanto numéricas como categóricas:
- **Variable Numérica**: la antigüedad del coche (en años), el kilometraje.
- **Variable Categórica**: el tipo de combustible (gasolina, diésel, eléctrico), el color del coche.

Para ajustar el modelo de regresión lineal, primero se transforma la variable categórica, por ejemplo, convirtiendo el tipo de combustible en variables ficticias como "gasolina=1, diésel=0, eléctrico=0". Luego, el modelo puede utilizar tanto las variables numéricas como estas variables ficticias para estimar el precio del coche.

La distinción entre variables numéricas y categóricas es fundamental, ya que define cómo se preprocesan y cómo se interpretan los resultados del modelo. Es importante tener en cuenta la naturaleza de las variables para elegir las técnicas de transformación y ajustes adecuados para construir modelos precisos.

## Regresiones y sus tipos
La regresión es una técnica estadística utilizada para modelar la relación entre una variable dependiente y una o más variables independientes. Es una herramienta fundamental para predecir valores y comprender cómo cambian ciertos fenómenos cuando otros varían. Existen diferentes tipos de regresión, cada una con características y aplicaciones específicas.

### Regresión Lineal Simple
La regresión lineal simple se utiliza cuando queremos modelar la relación entre una variable dependiente y una sola variable independiente. Esta relación se representa mediante una línea recta que se ajusta para minimizar la diferencia entre los valores observados y los valores predichos. Por ejemplo, podemos utilizar la regresión lineal simple para predecir el precio de una casa basándonos en su área en metros cuadrados.

### Regresión Lineal Múltiple
La regresión lineal múltiple es una extensión de la regresión lineal simple que permite modelar la relación entre una variable dependiente y más de una variable independiente. Es útil cuando se desea analizar cómo diferentes factores afectan una variable de respuesta. Por ejemplo, podríamos predecir el precio de una casa teniendo en cuenta varias variables como el área, la ubicación, el número de habitaciones, entre otras.

### Regresión Logística Simple
La regresión logística simple se utiliza cuando la variable dependiente es categórica binaria (es decir, solo tiene dos posibles resultados, como "sí" o "no", "positivo" o "negativo"). La regresión logística no modela una relación lineal, sino que estima la probabilidad de que ocurra un evento, utilizando una función sigmoide. Un ejemplo de regresión logística simple podría ser predecir si un cliente comprará un producto o no, basándonos en el tiempo que pasa mirando el producto.

### Regresión Logística Múltiple
La regresión logística múltiple se emplea cuando la variable dependiente es binaria y hay más de una variable independiente. En este caso, se utiliza para analizar el efecto de varias variables explicativas en la probabilidad de que ocurra un evento determinado. Por ejemplo, podríamos predecir si una persona tiene una enfermedad en función de varias variables como la edad, el índice de masa corporal, los hábitos alimenticios, etc.

### Mapa Conceptual de los Tipos de Regresión

A continuación, un resumen en forma de un esquema simple de los tipos de regresión descritos:

- **Regresión Estadística**
  - **Regresión Lineal**
    - Regresión Lineal Simple (1 variable independiente)
    - Regresión Lineal Múltiple (Más de 1 variable independiente)
  - **Regresión Logística**
    - Regresión Logística Simple (Variable dependiente binaria, 1 variable independiente)
    - Regresión Logística Múltiple (Variable dependiente binaria, varias variables independientes)

La elección entre estos tipos de regresión depende del tipo de datos y del problema que estemos tratando de resolver. Mientras que la regresión lineal es adecuada para problemas de predicción de valores continuos, la regresión logística es ideal para clasificar resultados en categorías binarias.

## Método de los Minimos Cuadrados
### El Método de los Mínimos Cuadrados

El método de los mínimos cuadrados es una técnica matemática utilizada para encontrar la mejor aproximación posible a una serie de datos observados, ajustando una función (generalmente una línea recta) que minimice la suma de los errores al cuadrado. En el contexto de la regresión lineal, este método se emplea para determinar los coeficientes de la ecuación de la línea recta que mejor se ajusta a los datos. Su objetivo es reducir al mínimo las diferencias entre los valores observados y los valores predichos por el modelo.

En la regresión lineal simple, la relación entre una variable dependiente ($y$) y una variable independiente ($x$) se modela con una línea recta de la forma:

\[ $y = b_0 + b_1x$ \]

donde:
- **\( $b_0$ \)** es la intersección con el eje y, es decir, el valor de \( y \) cuando \( x \) es cero.
- **\( $b_1$ \)** es la pendiente de la línea, que representa la tasa de cambio de \( y \) con respecto a \( x \).

El método de los mínimos cuadrados busca encontrar los valores de \( $b_0$ \) y \( $b_1$ \) que minimicen la suma de los cuadrados de las diferencias (errores) entre los valores observados ( $y_i$ ) y los valores predichos( $ŷ_i$). Estos errores se llaman "residuos" y la función objetivo que se minimiza es la siguiente:

$$
S = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \sum_{i=1}^{n} (y_i - (b_0 + b_1 x_i))^2
$$

donde:
- **\( $y_i$ \)** es el valor real de la variable dependiente en la observación \(i\).
- **\( $ŷ_i$ \)** es el valor predicho por el modelo para la observación \(i\).
- **\( $n$ \)** es el número total de observaciones.

### Cómo Funciona el Método de los Mínimos Cuadrados
El proceso de mínimos cuadrados se basa en minimizar esta suma de los errores al cuadrado. Esto implica que el modelo intenta ajustar los parámetros de tal manera que la diferencia total entre los valores observados y los valores predichos sea lo menor posible. Los pasos básicos incluyen:

1. **Calcular los Residuos**: Para cada observación, se calcula la diferencia entre el valor observado y el valor predicho por el modelo.
2. **Elevar al Cuadrado los Residuos**: Para evitar que los errores positivos y negativos se cancelen entre sí, se eleva cada residuo al cuadrado.
3. **Minimizar la Suma de los Cuadrados**: Se ajustan los coeficientes \( b_0 \) y \( b_1 \) de tal manera que la suma de los residuos al cuadrado sea mínima.

### Aplicación del Método de los Mínimos Cuadrados
El método de los mínimos cuadrados es fundamental en la construcción de modelos de regresión lineal debido a su simplicidad y eficacia para encontrar la mejor línea de ajuste. Al minimizar la suma de los errores al cuadrado, se asegura que los errores grandes tengan más peso, lo cual resulta en una línea ajustada que tiende a ser una buena representación del conjunto de datos.

En el ejemplo de predicción de salarios, el método de los mínimos cuadrados se utilizaría para determinar la línea que mejor describe la relación entre los años de experiencia y el salario. De esta manera, el modelo intenta minimizar las diferencias entre los salarios reales de los empleados y los salarios predichos por el modelo.

### Limitaciones del Método de los Mínimos Cuadrados
Aunque el método de los mínimos cuadrados es eficaz, tiene algunas limitaciones. Por ejemplo, es sensible a valores atípicos. Si hay observaciones extremas en el conjunto de datos, estas pueden tener un impacto significativo en la línea ajustada, lo que puede distorsionar los resultados. Además, la suposición de una relación lineal entre las variables no siempre se cumple, y en tales casos se requieren modelos más complejos.

A pesar de estas limitaciones, el método de los mínimos cuadrados sigue siendo una herramienta esencial para la regresión lineal, proporcionando una base sólida para comprender la relación entre variables y hacer predicciones con una interpretación clara.

En este caso práctico:

### Aplicación del Modelo

Con los datos disponibles sobre "Años de Experiencia" y "Salario", se puede entrenar un modelo de regresión lineal que encuentre la ecuación de una línea recta que mejor se ajuste a estos datos. La ecuación del modelo tiene la forma:

\[ $y = b_0 + b_1 × x$ \]

donde:
- \( $y$ \) es el salario predicho.
- \( $x$ \) son los años de experiencia.
- \( $b_0$ \) es el valor del salario cuando los años de experiencia son cero (la intersección con el eje y).
- \( $b_1$ \) es la pendiente de la línea, que indica cuánto cambia el salario por cada año adicional de experiencia.

Por ejemplo, si entrenamos el modelo con los datos proporcionados y obtenemos una pendiente de \( b_1 = 9500 \), esto significaría que cada año adicional de experiencia se asocia, en promedio, con un aumento de $9500 en el salario del empleado.

- **Evaluaciones y Promociones**: Ayuda a tomar decisiones salariales más equitativas y consistentes en función de la experiencia.
