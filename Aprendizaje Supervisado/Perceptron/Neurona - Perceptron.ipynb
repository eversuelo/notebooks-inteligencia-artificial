# Definicin formal de una Neurona Artificial
### Definicin Formal de una Neurona Artificial

Una **neurona artificial** es el componente fundamental de una red neuronal artificial, inspirado en las neuronas biolgicas. Su funcin principal es recibir varias entradas, combinarlas y producir una salida que depende de la suma ponderada de dichas entradas. Esta neurona imita el proceso biolgico mediante el cual una neurona en el cerebro recibe estmulos y decide si debe transmitir un impulso nervioso.

En trminos matemticos, la neurona artificial se puede describir utilizando un **modelo de entrada y salida** donde cada entrada est asociada a un peso. Esta combinacin ponderada se convierte en una **entrada neta**, que luego se pasa a una **funcin de activacin** que decide si la neurona se activa (dispara) o no.

La ecuacin de la entrada neta se define de la siguiente manera:

$$z = w_1x_1 + w_2x_2 + \dots + w_mx_m + b$$

donde:
- \( $z$ \) es la **entrada neta**,
- \( $w_i$ \) es el **peso** asociado a la entrada \( x_i \),
- \( $x_i$ \) son los **valores de entrada**, y
- \( $b$ \) es el **sesgo**.

La funcin de activacin \(\sigma(z)\) toma la entrada neta y genera la salida de la neurona. En el caso de un **problema de clasificacin binaria** con dos clases, 0 y 1, la funcin de activacin puede ser una funcin escaln unitario:

$$
\sigma(z) = 
\begin{cases}
1 & \text{si } z \geq 0 \\
0 & \text{si } z < 0
\end{cases}
$$


La idea detrs de la **neurona artificial** es combinar las entradas, ponderadas por los pesos, para determinar la salida. Esto es similar a cmo una neurona biolgica suma los potenciales que recibe a travs de sus dendritas. Si la suma alcanza un cierto umbral, la neurona se activa y transmite un impulso elctrico. Del mismo modo, en la neurona artificial, si la entrada neta es mayor o igual que un cierto umbral (en este caso, cero), la salida ser 1; de lo contrario, ser 0.

#### Expansin Matemtica: El Sesgo y el Producto Escalar
Para simplificar la representacin, el umbral \($\theta$\) se puede mover al lado izquierdo de la ecuacin y redefinir como un **sesgo** (\($b$\)). Esto nos da una expresin simplificada:

$$z = \mathbf{w}^T \mathbf{x} + b$$

donde \(\mathbf{w}^T\) es el vector transpuesto de pesos y \(\mathbf{x}\) es el vector de entradas. Esta formulacin con el producto escalar facilita la implementacin de la neurona en trminos computacionales, ya que se puede generalizar para incluir mltiples entradas.

La funcin de activacin, adems, no se limita a ser una funcin escaln. Algunas de las funciones ms populares son:
- **Sigmoide**: \($\sigma(z) = \frac{1}{1 + e^{-z}}$\), que genera una salida en el rango de 0 a 1, til para problemas de clasificacin donde se desea una probabilidad.
- **ReLU (Rectified Linear Unit)**: \($\text{ReLU}(z) = \max(0, z)$\), que es muy popular en redes neuronales profundas debido a su capacidad de introducir no linealidades sin saturarse para valores grandes de entrada.

### Aplicaciones de la Neurona Artificial
Las neuronas artificiales se utilizan como bloques de construccin en redes ms complejas conocidas como **redes neuronales artificiales**. Cuando se combinan mltiples neuronas en varias capas (redes neuronales multicapa), se pueden abordar problemas ms complejos, como la clasificacin de imgenes, el procesamiento del lenguaje natural, y muchas otras aplicaciones del aprendizaje profundo.

En resumen, la **neurona artificial** es un componente esencial que permite a las computadoras aprender de los datos, ajustando los pesos y el sesgo a travs de algoritmos como el **descenso de gradiente** para reducir el error en las predicciones. Esta simple unidad, cuando se agrupa con otras en una arquitectura de red, se convierte en una poderosa herramienta capaz de aprender patrones complejos y realizar predicciones precisas.

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = 0

    def step_function(self, x):
        return 1 if x >= 0 else 0

    def fit(self, X, y):
        # Inicializar pesos
        self.weights = np.zeros(X.shape[1])

        # Iterar sobre el nmero de pocas
        for epoch in range(self.epochs):
            for xi, target in zip(X, y):
                # Calcular la prediccin
                linear_output = np.dot(xi, self.weights) + self.bias
                y_pred = self.step_function(linear_output)
                
                # Actualizar pesos y bias
                update = self.learning_rate * (target - y_pred)
                self.weights += update * xi
                self.bias += update

            # Imprimir pesos y bias despus de cada poca
            print(f"poca {epoch+1}: w = {self.weights}, bias = {self.bias}")

    def predict(self, X):
        # Aplicar el perceptrn a nuevas entradas
        linear_output = np.dot(X, self.weights) + self.bias
        return np.array([self.step_function(x) for x in linear_output])

# Datos de entrenamiento (Ejemplo con AND lgico)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Entrenar el perceptrn
perceptron = Perceptron(learning_rate=0.1, epochs=10)
perceptron.fit(X, y)

# Prueba del modelo
print("Predicciones:", perceptron.predict(X))
```
